```python
#ex 3: api with flask
import flask
from flask import request, jsonify

app = flask.Flask(__name__)

# products list
products = [
    {'id': 0,'name': 'banana','description': 'banana','amount': '100','unit_price': '20'},
    {'id': 1,'name': 'mango','description': 'mango','amount': '60','unit_price': '25'},
    {'id': 2,'name': 'apple','description': 'apple','amount': '50','unit_price': '30'},
    {'id': 3,'name': 'orange','description': 'orange','amount': '100','unit_price': '100'},
]


@app.route('/', methods=['GET'])
def home():
    return '''<h1>Flask Friut shop</h1>
<p>simple Flask friut shop API </p>'''


@app.route('/api/v1/product/all', methods=['GET'])
def getAll():
    return jsonify(products)


@app.route('/api/v1/product/id', methods=['GET'])
def getProductByID():
    
    if 'id' in request.args:
        pid = int(request.args['id'])
    else:
        return "Product not found . Please specify an id."

    results = []


    for product in products:
        if product['id'] == pid:
            results.append(product)
            
    return jsonify(results)

@app.route('/api/v1/product/name', methods=['GET'])
def getProductByName():
    
    if 'name' in request.args:
        pname = request.args['name']
    else:
        return "Product not found . Please specify a product's name."

    results = []
    for product in products:
        if product['name'] == pname:
            results.append(product)
            
    return jsonify(results)

@app.errorhandler(404)
def page_not_found(e):
    return "<h1>404</h1><p>Page could not be found.</p>", 404

if __name__ == '__main__':
    print('Flask friut shop  app started !!')
    app.run(port = 5006)
```

```python
## Flask app with advance qrcode
## response to qrcoder.html
from flask import Flask, request, send_file
from flask_cors import CORS

import qrcode
from io import BytesIO


app = Flask(__name__)
CORS(app)

@app.route('/generate_qrcode', methods=['POST'])
def generateQrcode():
    buffer = BytesIO()
    data = request.form.get('data')   
    img = qrcode.make(data)
    img.save(buffer)
    buffer.seek(0)
    
    resp = send_file(buffer, mimetype='image/png')
    return resp


if __name__== '__main__':
    app.run(port=5005)


```

```python
## Flask app with advance qrcode
##  response to qrcoder2.html
from flask import Flask, request, send_file
from flask_cors import CORS

import qrcode
from io import BytesIO


app = Flask(__name__)
CORS(app)

@app.route('/generate_qrcode', methods=['POST'])
def generateQrcode():
    buffer = BytesIO()
    data = request.form.get('data')   
    version = int(request.form.get('version'))  
    box_size = int(request.form.get('box_size'))  
    box_border = int(request.form.get('box_border')) 
    fillColor = request.form.get('fill_color')
    backColor = request.form.get('back_color')    
    qr = qrcode.QRCode(
    version=version,
    #error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=box_border,
    border=box_border,
    )
    qr.add_data(data)
    qr.make(fit=True)
    
    img = qr.make_image(fill_color=fillColor, back_color=backColor)
    #img = qrcode.make(data)
    img.save(buffer)
    buffer.seek(0)
    
    resp = send_file(buffer, mimetype='image/png')
    return resp


if __name__== '__main__':
    app.run(port=5006)
```
